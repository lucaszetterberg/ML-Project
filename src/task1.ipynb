```python
import pylab as pb
import numpy as np
from math import pi
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
## Task 1.1 (set the prior distribution over W and visualise it)

# Define the prior mean and covariance
prior_mean = np.zeros(2)
prior_covariance = 0.2 * np.eye(2)
alpha = 0.2

# Sample from the prior distribution

samplesize = 5000

# Generate samples from the prior distribution

prior = np.random.multivariate_normal(prior_mean, prior_covariance, size= samplesize)

# Extract the first and second dimension of the samples
prior1 = prior[:, 0]
prior2 = prior[:, 1]

# Plot the samples
plt.figure()
plt.scatter(prior1,prior2,alpha)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Prior distribution over W')
plt.show()


```

```python
## task 1.2 (Pick a single data point (x,t) and visualise the posterior distribution over W)

# Define the data point

x = 0.5
t = 0.5


## calculate the likelihood

likelihood = np.exp(-0.5 * (t - prior1 * x) ** 2)



# Extract the first and second dimension of the samples

posterior1 =  likelihood * prior[:, 0]
posterior2 = likelihood * prior[:, 1]



# Plot the samples
plt.figure()
plt.scatter(posterior1,posterior2,alpha)
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Posterior distribution over W')
plt.show()





    
```
