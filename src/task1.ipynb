```python
import pylab as pb
import numpy as np
from math import pi
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
# Define the prior mean and covariance
prior_mean = np.zeros(2)
prior_covariance = 0.2 * np.eye(2)
alpha = 0.2

# Sample from the prior distribution

samplesize = 5000

f = np.random.multivariate_normal(prior_mean, prior_covariance, size= samplesize)

f1 = f[:, 0]
f2 = f[:, 1]

plt.figure()
plt.scatter(f1,f2,alpha)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Prior distribution over W')
plt.show()


```
